```python
from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
from shutil import copyfile
```

```python
data_dir = Path("../data/dataset")
```

```python
image_list = list(map(str, (data_dir / "train").rglob("*.jpg")))
dir_list = set('/'.join(x.split("/")[:6]) for x in image_list)
one_image_list = [list(Path(x_dir).glob('*.jpg'))[0] for x_dir in dir_list]
one_image_list = [str(x) for x in one_image_list]
one_mask_list = [x[:-4]+".png" for x in one_image_list]
one_trimap_list = [x[:-4]+"_trimap_true.png" for x in one_image_list]
```

```python
one_trimap_list
```

```python
for i, src_img_path in enumerate(one_image_list):
    target_dir = "../data/one_image_dataset/" + '/'.join(src_img_path.split("/")[3:6]) + '/'    
    Path(target_dir).mkdir(parents=True, exist_ok=True)
    copyfile(src_img_path, target_dir+src_img_path.split('/')[-1])
    print(target_dir+one_mask_list[i].split('/')[-1])
    copyfile(one_mask_list[i], target_dir+one_mask_list[i].split('/')[-1])
    copyfile(one_trimap_list[i], target_dir+one_trimap_list[i].split('/')[-1])
```
