```python
from pathlib import Path
import numpy as np
import shutil
```

```python
train_size = 170
validation_size = 30
test_size = 20
```

```python
src_data_dir = "../data/dataset/train"
tgt_data_dir = "../data/dataset_split"
```

```python
dir_list = [str(x) for x in Path(src_data_dir).glob('**/*') if x.is_dir()][2:]
```

```python
img_list = list(map(str, Path(src_data_dir).rglob("*.*")))
```

```python
dir_list
```

```python
ids = np.arange(0, 220)
np.random.shuffle(ids)
```

```python
train_ids = ids[:train_size]
val_ids = ids[train_size:train_size+validation_size]
test_ids = ids[train_size+validation_size:]
```

```python
train_ids, val_ids, test_ids
```

```python
len(dir_list)
```

```python
train_dirs, val_dirs, test_dirs = (np.array(dir_list)[train_ids],
    np.array(dir_list)[val_ids], np.array(dir_list)[test_ids])
```

```python
#for train_dir in train_dirs:
#    shutil.copytree(str(train_dir), tgt_data_dir + '/train/' + '/'.join(train_dir.split('/')[4:]))

for val_dir in val_dirs:
    shutil.copytree(str(val_dir), tgt_data_dir + '/val/' + '/'.join(val_dir.split('/')[4:]))

for test_dir in test_dirs:
    shutil.copytree(str(test_dir), tgt_data_dir + '/test/' + '/'.join(test_dir.split('/')[4:]))
```
