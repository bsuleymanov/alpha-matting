```python
def mkdir_if_empty_or_not_exist(dir_name):
    if (not dir_name.exists() or
        next(dir_name.iterdir(), None) is None):
        Path.mkdir(dir_name, exist_ok=True)
    else:
        raise Exception
```

```python
h, w = np.array(random_mask).shape
np_boundary = np.zeros([h, w])

side = int((h + w) / 2 * 0.05)
dilated = grey_dilation(random_mask, size=(side, side))
eroded = grey_erosion(random_mask, size=(side, side))

np_boundary[np.where(dilated - eroded != 0)] = 255
```

```python
for mask_path in train_masks:
    mask = Image.open(mask_path)
    mask = np.array(mask)
    if len(mask.shape) > 2:
        mask = mask[..., 0]
    h, w = mask.shape
    np_boundary = np.zeros([h, w])

    side = int((h + w) / 2 * 0.05)
    dilated = grey_dilation(mask, size=(side, side))
    eroded = grey_erosion(mask, size=(side, side))

    np_boundary[np.where(dilated - eroded != 0)] = 1.
    
    np_boundary = (((np_boundary - np_boundary.min()) / (np_boundary.max() - np_boundary.min())) * 255.0).astype(np.uint8)
    Image.fromarray(np_boundary).save(mask_path[:-4]+"_trimap.png")
```

```python
for res_img_path in tqdm.tqdm(res_images):
    #print(res_img_path)
    img = Image.open(res_img_path)
    if img.mode == "RGB":
        img = np.array(img)
        img = np.mean(img, axis=2)
        img = Image.fromarray(img).convert("L")
    save_path = "../src/result_submission/"+'/'.join(res_img_path.split('/')[3:])
    Path(save_path).parents[0].mkdir(exist_ok=True, parents=True)
    img.save(save_path, "PNG")
```
